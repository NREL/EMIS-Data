```python
import pandas as pd
```

```python
def calc_on_time(uc):
    diff = uc.diff()
 
    initial_state_dict = dict()
    for col in diff.columns:
        s = diff[col].dropna()
        diff_s = s[s!=0]
        if diff_s.empty:
            check_row = uc[col].head(1)
        else:
            check_row = diff_s.tail(1)
 
        if check_row.values == -1 or check_row.values == 0:
            initial_state_dict[col] = -1 * (len(uc.index) - uc.index.get_loc(check_row.index[0]))
        else:
            initial_state_dict[col] = len(uc.index) - uc.index.get_loc(check_row.index[0])
            
    return(initial_state_dict)
```

```python
uc = pd.read_csv('PLEXOS_DA_solution_commitment.csv',index_col='time')
```

```python
uc
```

```python
on_time = calc_on_time(uc.loc[uc.index < '2020-07-12 00:00:00'])
```

```python
import csv

with open('on_time.csv', 'wb') as f:  # Just use 'w' mode in 3.x
    w = csv.DictWriter(f, on_time.keys())
    w.writeheader()
    w.writerow(on_time)
```
